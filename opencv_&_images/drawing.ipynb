```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
plt.imshow(blank_img)

# draw a rectangle
drawn = cv2.rectangle(blank_img, (100, 0), (300, 300), (0, 255, 0), 10)
plt.imshow(drawn)
```

```python
cir_img = cv2.circle(img=blank_img, center=(250, 250), radius=50, color=(255, 0, 0), thickness=5)

plt.imshow(cir_img)
```

```python
line_img = cv2.line(blank_img, pt1=(20, 20), pt2=(500, 500), color=(0, 0, 255), thickness=5)

plt.imshow(line_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
wt = cv2.putText(img=blank_img,text="hello", org=(10, 500), fontFace=font, fontScale=1, color=(255,255,255), thickness=2)
plt.imshow(wt)
```

```python
blank_img2 = np.zeros(shape=(512, 512, 3), dtype=np.int32)
plt.imshow(blank_img2)

vertices = np.array([[100, 300],[200, 200],[400, 300],[200, 400]], dtype=np.int32)
pts = vertices = vertices.reshape((-1, 1, 2))

new_i = cv2.polylines(blank_img2, [pts], isClosed=True, color=(0, 255, 0), thickness=5)
plt.imshow(new_i)
```
