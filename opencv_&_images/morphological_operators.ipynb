```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# why to use morphological operators in computer vision apps?
# It helps with feature separation and extraction 
```

```python
def loadImage():
    img = np.zeros((600, 600))
    font = cv2.FONT_HERSHEY_COMPLEX
    img = cv2.putText(img, text="ABCDE", org=(50,300), fontFace=font, fontScale=5, color=(255, 255, 255), thickness=15)
    return img
```

```python
def display_image(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
img = loadImage()
display_image(img)
```

```python
img = loadImage()
kernal = np.ones((5, 5), dtype=np.uint8)
result = cv2.erode(img, kernal, iterations=3)
display_image(result)
```

```python
img = loadImage()
white_noise = np.random.randint(low=0, high=2, size=(600, 600))
# display_image(white_noise)
white_noise = white_noise * 255
noise_img = white_noise + img
display_image(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernal)
display_image(opening)
```

```python
img = loadImage()
black_noise = np.random.randint(low=0, high=2,size=(600, 600))
black_noise = black_noise * -255
black_noise_image = img + black_noise
display_image(black_noise_image)
```
