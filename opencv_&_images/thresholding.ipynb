```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the image in as gray scale
img = cv2.imread("../DATA/rainbow.jpg", 0)
plt.imshow(img)
```

```python
# creates a threshold, pixels that have a value above 127 become 255 (inverse), non inverse threshold param is something too
ret,thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh1, cmap="gray")
```

```python
img2 = cv2.imread("../DATA/crossword.jpg", 0)
plt.imshow(img2, cmap="gray")
```

```python
def showPic(img):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
showPic(img2)
```

```python
ret, th1 = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)
showPic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
showPic(th2)
```

```python
# combine different thresholds for a clearer grayscale image
bld = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
showPic(bld)
```
