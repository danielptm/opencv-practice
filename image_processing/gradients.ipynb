```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# Why use gradients in computer vision apps?
# It helps to reveal textures, edges and other important features
```

```python
def loadImage():
    img = np.zeros((600, 600))
    font = cv2.FONT_HERSHEY_COMPLEX
    img = cv2.putText(img, text="ABCDE", org=(50,300), fontFace=font, fontScale=5, color=(255, 255, 255), thickness=15)
    return img
```

```python
def display_image(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
img = cv2.imread("../DATA/sudoku.jpg", 0)
display_image(img)
```

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0,ksize=5)
display_image(sobelx)
```

```python
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1,ksize=5)
display_image(sobely)
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
display_image(laplacian)
```

```python
blended = cv2.addWeighted(src1=sobelx, src2=sobely, alpha=-0.5, beta=0.5, gamma=0)
display_image(blended)
```

```python
ret, th1 = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
display_image(th1)
```

```python
kernel = np.ones((4, 4), dtype=np.uint8)
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT, kernel)
display_image(gradient)
```
