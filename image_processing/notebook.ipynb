```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
# reads into numpy array
img = cv2.imread("../DATA/00-puppy.jpg")
type(img)

img.shape

# opencv reads the image as BGR rather than RGB so the image looks weird.
plt.imshow(img)

# You can fix the color conversion like this
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)

# You can also read it in with the conversion
gray = cv2.imread("../DATA/00-puppy.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(gray, cmap="gray")

# resize image with absolute numnbers, resize by ratio is more common
new_img = cv2.resize(img2, (1000, 400))
plt.imshow(new_img)

# flip image
ni = cv2.flip(img2, 1)
plt.imshow(ni)



type(img)
# Save image
# cv2.imwrite('totally_newjpeg.jpg', img)

```
