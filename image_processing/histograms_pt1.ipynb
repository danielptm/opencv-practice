```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread("../DATA/horse.jpg")
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread("../DATA/rainbow.jpg")
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

bricks = cv2.imread("../DATA/bricks.jpg")
show_bricks = cv2.cvtColor(bricks, cv2.COLOR_BGR2RGB)

plt.imshow(show_bricks)
```

```python
hist_values = cv2.calcHist([bricks], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
hist_values.shape
plt.plot(hist_values)
```

```python
hist_values = cv2.calcHist([dark_horse], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
hist_values.shape
plt.plot(hist_values)
```

```python
hist_values = cv2.calcHist([bricks], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
hist_values.shape
plt.plot(hist_values)
```

```python
img = bricks
color = ('b', 'g', 'r')
for i,color in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color=color)
    plt.xlim([0,256])
plt.title("Historgram for bricks")
```

```python
img = dark_horse
color = ('b', 'g', 'r')
for i,color in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color=color)
    plt.xlim([0,50])
    plt.ylim([0, 500000])
plt.title("Historgram for horse")
```
